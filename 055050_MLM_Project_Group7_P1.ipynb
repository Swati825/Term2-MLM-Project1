```python
# Handling Missing Data
def missing_data_summary(df):
    """
    Creates a DataFrame summarizing missing data statistics.

    Args:
        df: The input DataFrame.

    Returns:
        A DataFrame with missing data count and percentage for each variable and row.
    """
    # Column-wise summary
    missing_count_col = df.isnull().sum()
    missing_percentage_col = (missing_count_col / len(df)) * 100
    missing_summary_col = pd.DataFrame({
        'Missing Count': missing_count_col,
        'Missing Percentage': missing_percentage_col
    })
    # Row-wise summary
    missing_count_row = df.isnull().sum(axis=1)
    missing_percentage_row = (missing_count_row / len(df.columns)) * 100
    missing_summary_row = pd.DataFrame({
        'Missing Count': missing_count_row,
        'Missing Percentage': missing_percentage_row
    }, index=df.index)  # Use original DataFrame index

    return missing_summary_col, missing_summary_row


# Create categorical and non-categorical subsets
rs35_ds_cat = rs35_ds[rs35_cat_var]
rs35_ds_noncat = rs35_ds[rs35_noncat_var]

# Calculate missing data summary for categorical subset
rs35_missing_summary_cat_col, rs35_missing_summary_cat_row = missing_data_summary(rs35_ds_cat)

# Calculate missing data summary for non-categorical subset
rs35_missing_summary_noncat_col, rs35_missing_summary_noncat_row = missing_data_summary(rs35_ds_noncat)

# Drop columns and rows with more than 50% missing data for categorical subset
rs35_ds_cat_treated = rs35_ds_cat.dropna(thresh=len(rs35_ds_cat) * 0.5, axis=1)
rs35_ds_cat_treated = rs35_ds_cat_treated.dropna(thresh=len(rs35_ds_cat_treated.columns) * 0.5, axis=0)

# Drop columns and rows with more than 50% missing data for non-categorical subset
rs35_ds_noncat_treated = rs35_ds_noncat.dropna(thresh=len(rs35_ds_noncat) * 0.5, axis=1)
rs35_ds_noncat_treated = rs35_ds_noncat_treated.dropna(thresh=len(rs35_ds_noncat_treated.columns) * 0.5, axis=0)

# Calculate total variables and rows dropped
rs35_dropped_variables_cat = len(rs35_ds_cat.columns) - len(rs35_ds_cat_treated.columns)
rs35_dropped_rows_cat = len(rs35_ds_cat) - len(rs35_ds_cat_treated)
rs35_dropped_variables_noncat = len(rs35_ds_noncat.columns) - len(rs35_ds_noncat_treated.columns)
rs35_dropped_rows_noncat = len(rs35_ds_noncat) - len(rs35_ds_noncat_treated)

rs35_total_dropped_variables = rs35_dropped_variables_cat + rs35_dropped_variables_noncat
rs35_total_dropped_rows = rs35_dropped_rows_cat + rs35_dropped_rows_noncat

# Print missing data summaries
print("Categorical Missing Data Summary (Columns):")
print(rs35_missing_summary_cat_col)
print("\nCategorical Missing Data Summary (Rows):")
print(rs35_missing_summary_cat_row)
print("\nNon-Categorical Missing Data Summary (Columns):")
print(rs35_missing_summary_noncat_col)
print("\nNon-Categorical Missing Data Summary (Rows):")
print(rs35_missing_summary_noncat_row)

# Print total dropped variables and rows
print(f"\nTotal variables dropped: {rs35_total_dropped_variables}")
print(f"Total rows dropped: {rs35_total_dropped_rows}")
```

**Numerical Encoding of Non-Numeric Categorical Data**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder

# Assume 'my_ds_name' is the DataFrame
my_ds_name = rs35_ds.copy()

# List of categorical columns
categorical_cols = ['Country', 'Product', 'Import_Export', 'Category', 'Port',
                    'Customs_Code', 'Shipping_Method', 'Supplier', 'Customer', 'Payment_Terms']

# Filter existing columns
existing_cols = [col for col in categorical_cols if col in my_ds_name.columns]

# Pandas Encoding
ds_cat_mdt_code_pd = my_ds_name.copy()
for col in existing_cols:
    ds_cat_mdt_code_pd[col] = ds_cat_mdt_code_pd[col].astype('category')
    ds_cat_mdt_code_pd[col + '_code'] = ds_cat_mdt_code_pd[col].cat.codes

# Scikit-learn LabelEncoder
ds_cat_mdt_code_le = my_ds_name.copy()
le = LabelEncoder()
for col in existing_cols:
    ds_cat_mdt_code_le[col + '_code'] = le.fit_transform(ds_cat_mdt_code_le[col])

# Scikit-learn OrdinalEncoder
ds_cat_mdt_code_oe = my_ds_name.copy()
oe = OrdinalEncoder()
encoded_cols = [col + '_code' for col in existing_cols]
ds_cat_mdt_code_oe[encoded_cols] = oe.fit_transform(ds_cat_mdt_code_oe[existing_cols])

# Verify results
print(ds_cat_mdt_code_pd.head())
print(ds_cat_mdt_code_le.head())
print(ds_cat_mdt_code_oe[encoded_cols].head())
```

```python
import pandas as pd
from scipy.stats import spearmanr, kendalltau

# Function to calculate descriptive statistics for categorical variables
def rs35_categorical_stats(df, cat_columns):
    rs35_results = []

    for rs35_col in cat_columns:
        rs35_counts = df[rs35_col].value_counts()
        rs35_proportions = rs35_counts / len(df)
        rs35_mode = rs35_counts.idxmax()
        rs35_min = df[rs35_col].min()
        rs35_max = df[rs35_col].max()

        for category, count in rs35_counts.items():
            rs35_results.append({
                "Column": rs35_col,
                "Category": category,
                "Frequency": count,
                "Proportion": rs35_proportions[category],
                "Minimum": rs35_min,
                "Maximum": rs35_max,
                "Mode": rs35_mode
            })

    return pd.DataFrame(rs35_results)

# Function to calculate correlation for categorical columns
def rs35_categorical_correlation(df, cat_columns, method='spearman'):
    rs35_corr_results = pd.DataFrame(index=cat_columns, columns=cat_columns, dtype=float)

    for i in range(len(cat_columns)):
        for j in range(i, len(cat_columns)):
            rs35_col1, rs35_col2 = cat_columns[i], cat_columns[j]
            if i == j:
                # Perfect correlation for the same variable
                corr = 1.0
            else:
                if method == 'spearman':
                    corr, _ = spearmanr(df[rs35_col1].astype('category').cat.codes,
                                        df[rs35_col2].astype('category').cat.codes)
                elif method == 'kendall':
                    corr, _ = kendalltau(df[rs35_col1].astype('category').cat.codes,
                                         df[rs35_col2].astype('category').cat.codes)
            rs35_corr_results.loc[rs35_col1, rs35_col2] = corr
            rs35_corr_results.loc[rs35_col2, rs35_col1] = corr

    return rs35_corr_results

# Retrieve categorical columns from rs35_ds
rs35_cat_columns = rs35_cat_var  # List of categorical variables

# Generate descriptive stats and correlations for these columns
rs35_categorical_stats_df = rs35_categorical_stats(rs35_ds, rs35_cat_columns)
rs35_spearman_corr_df = rs35_categorical_correlation(rs35_ds, rs35_cat_columns, method='spearman')
rs35_kendall_corr_df = rs35_categorical_correlation(rs35_ds, rs35_cat_columns, method='kendall')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
rs35_spearman_corr_df.fillna(1, inplace=True)
rs35_kendall_corr_df.fillna(1, inplace=True)

# Display results
print("Descriptive Statistics:")
print(rs35_categorical_stats_df)  # Show all rows

print("\nSpearman Correlation:")
print(rs35_spearman_corr_df)

print("\nKendall Correlation:")
print(rs35_kendall_corr_df)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr, skew, kurtosis, variation
from statsmodels.stats.weightstats import DescrStatsW


def descriptive_stats_non_categorical(rs35_df, non_categorical_columns):
    """
    Calculates descriptive statistics for non-categorical variables.

    Args:
        df (pd.DataFrame): The input DataFrame.
        non_categorical_columns (list): List of non-categorical column names.

    Returns:
        dict: A dictionary containing descriptive statistics for each column.
    """
    results = {}
    for column in non_categorical_columns:
      col_data = rs35_df[column].dropna()  # Exclude NaN values for calculations

      if not col_data.empty:
        min_val = col_data.min()
        max_val = col_data.max()
        mean_val = col_data.mean()
        median_val = col_data.median()
        mode_val = col_data.mode().iloc[0] if not col_data.mode().empty else None #Handle empty series for mode

        percentiles = {
            '25th': col_data.quantile(0.25),
            '50th': col_data.quantile(0.50),
            '75th': col_data.quantile(0.75)
        }
        range_val = max_val - min_val
        std_dev = col_data.std()
        skew_val = skew(col_data)
        kurt_val = kurtosis(col_data)

        results[column] = {
            'minimum': min_val,
            'maximum': max_val,
            'mean': mean_val,
            'median': median_val,
            'mode': mode_val,
            'percentiles': percentiles,
            'range': range_val,
            'standard_deviation': std_dev,
            'skewness': skew_val,
            'kurtosis': kurt_val
            }
    return results


def correlation_non_categorical(rs35_df, non_categorical_columns, method='pearson'):
    """
    Calculates correlation for non-categorical variables.

    Args:
        df (pd.DataFrame): The input DataFrame.
        non_categorical_columns (list): List of non-categorical column names.
        method (str): 'pearson' or 'spearman'.

    Returns:
        dict: A dictionary containing correlation for each pair of columns.
    """
    correlations = {}
    for i in range(len(non_categorical_columns)):
        for j in range(i + 1, len(non_categorical_columns)):
            col1 = non_categorical_columns[i]
            col2 = non_categorical_columns[j]
            #Handling Nan Values
            df_subset = rs35_df[[col1, col2]].dropna()
            if not df_subset.empty: #To avoid correlation calulation on empty dataframe
                if method == 'pearson':
                    corr, _ = pearsonr(df_subset[col1], df_subset[col2])
                elif method == 'spearman':
                    corr, _ = spearmanr(df_subset[col1], df_subset[col2])
                else:
                    raise ValueError("Invalid method. Choose 'pearson' or 'spearman'.")

                correlations[(col1, col2)] = corr
    return correlations

def composite_measures(df, non_categorical_columns, confidence_level=0.95):
    """
    Calculates composite measures for non-categorical variables.

    Args:
        df (pd.DataFrame): The input DataFrame.
        non_categorical_columns (list): List of non-categorical column names.
        confidence_level: Confidence level for confidence intervals.

    Returns:
        dict: A dictionary containing composite measures for each column.
    """
    composite = {}
    for column in non_categorical_columns:
        col_data = rs35_df[column].dropna()
        if not col_data.empty:
            mean = col_data.mean()
            std = col_data.std()
            cv = variation(col_data) * 100
            dstats = DescrStatsW(col_data)
            ci = dstats.tconfint_mean(alpha=1-confidence_level)

            composite[column] = {
                'coefficient_of_variation': cv,
                'confidence_interval': ci
                }
    return composite


non_categorical_columns = ['Quantity', 'Value', 'Customs_Code', 'Weight', 'Invoice_Number']

# Calculate and print descriptive statistics
desc_stats = descriptive_stats_non_categorical(rs35_ds, non_categorical_columns)
for column, stats in desc_stats.items():
    print(f"\n--- Descriptive Statistics for {column} ---")
    for key, value in stats.items():
        print(f"{key}: {value}")


# Calculate and print Pearson correlation
pearson_corr = correlation_non_categorical(rs35_ds, non_categorical_columns, method='pearson')
print("\n--- Pearson Correlations ---")
for (col1, col2), corr in pearson_corr.items():
    print(f"Pearson correlation between {col1} and {col2}: {corr:.2f}")


# Calculate and print Spearman correlation
spearman_corr = correlation_non_categorical(rs35_ds, non_categorical_columns, method='spearman')
print("\n--- Spearman Correlations ---")
for (col1, col2), corr in spearman_corr.items():
    print(f"Spearman correlation between {col1} and {col2}: {corr:.2f}")


# Calculate and print composite measures
comp_measures = composite_measures(rs35_ds, non_categorical_columns)
print("\n--- Composite Measures ---")
for column, measures in comp_measures.items():
    print(f"\nComposite Measures for {column}:")
    for key, value in measures.items():
        print(f"{key}: {value}")
```

### Test of Homogeneity {Chi-sq}

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Example: Testing homogeneity between 'Import_Export' and 'Category'
# Ensure both variables are categorical
cat_var1 = 'Import_Export'  # Nominal variable
cat_var2 = 'Category'       # Nominal/Ordinal variable

# Create a contingency table
contingency_table = pd.crosstab(rs35_ds[cat_var1], rs35_ds[cat_var2])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-Square Test of Homogeneity
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Print Results
print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies Table:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("\nResult: Reject the null hypothesis (significant difference in proportions).")
else:
    print("\nResult: Fail to reject the null hypothesis (no significant difference in proportions).")
```

```python
# 3.1.1 Unsupervised Machine Learning: Clustering
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import silhouette_score

# Fill missing values in rs35_ds
for col in rs35_ds.columns:
    if rs35_ds[col].isnull().any():
        if pd.api.types.is_numeric_dtype(rs35_ds[col]):
            rs35_ds[col] = rs35_ds[col].fillna(rs35_ds[col].mean())
        else:
            rs35_ds[col] = rs35_ds[col].fillna(rs35_ds[col].mode()[0])

# Select numerical features for clustering
numerical_features = rs35_ds.select_dtypes(include=np.number).columns.tolist()

# Scale the numerical features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(rs35_ds[numerical_features])

# Elbow method to choose optimal k
inertia = []
for i in range(1, min(len(rs35_ds), 11)):  # Ensuring clusters don't exceed the number of samples
    kmeans = KMeans(n_clusters=i, random_state=35, n_init=10)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, min(len(rs35_ds), 11)), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.show()

# Silhouette score to choose optimal k
silhouette_scores = []
for i in range(2, min(len(rs35_ds), 11)):  # Ensuring clusters are valid for silhouette score
    kmeans = KMeans(n_clusters=i, random_state=35, n_init=10)
    kmeans.fit(scaled_features)
    if len(set(kmeans.labels_)) > 1:  # Ensure at least two distinct clusters
        silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))
    else:
        print(f"Skipping silhouette score for {i} clusters due to insufficient distinct labels.")

plt.plot(range(2, len(silhouette_scores) + 2), silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()

# Choose optimal number of clusters based on Elbow method or Silhouette score
optimal_k = 3  # Adjust this value based on your analysis of the plots

# Apply K-Means clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=35, n_init=10)
rs35_ds['cluster'] = kmeans.fit_predict(scaled_features)

# Final result with clustering assigned to a new DataFrame
rs35_ds_cluster = rs35_ds.copy()

# Analyze the clusters
print(rs35_ds_cluster['cluster'].value_counts())

# Profile the clusters
numerical_cols = rs35_ds_cluster.select_dtypes(include=np.number).columns
cluster_profiles = rs35_ds_cluster.groupby('cluster')[numerical_cols].mean()
print("Cluster Profiles (Numerical Data):")
print(cluster_profiles)

# If categorical features exist and need profiling
categorical_cols = [col for col in rs35_ds_cluster.columns if col not in numerical_features and col != 'cluster']
if categorical_cols:
    # Encode categorical columns
    rs35_ds_encoded = rs35_ds_cluster.copy()
    encoder = LabelEncoder()
    for col in categorical_cols:
        rs35_ds_encoded[col] = encoder.fit_transform(rs35_ds_encoded[col])

    # Profile clusters with encoded data
    cluster_profiles_categorical = rs35_ds_encoded.groupby('cluster').mean()
    print("Cluster Profiles (Encoded Data):")
    print(cluster_profiles_categorical)

# Visualize clusters using two numerical features
if len(numerical_features) >= 2:
    sns.scatterplot(
        x=numerical_features[0], y=numerical_features[1], hue='cluster',
        data=rs35_ds_cluster, palette='viridis', legend='full'
    )
    plt.title('Clusters Visualization')
    plt.show()

# Display the first few rows of the new DataFrame to confirm
print(rs35_ds_cluster.head())
```
